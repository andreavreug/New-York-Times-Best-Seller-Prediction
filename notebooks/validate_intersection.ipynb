# Process Data

#### Validate Intersection

```python
# Load datasets
import pandas as pd

# Explicit absolute paths (Windows) â€” adjust if your user/path differs
nyt_path = '../data/nyt_unique_books_2025.csv'
hc_path = '../data/Hardcover_2025.csv'

nyt = pd.read_csv(nyt_path)
hc = pd.read_csv(hc_path)

#Create book_id  (handles NaN and whitespace)
hc['book_id'] = (
    hc['authors'].fillna('').str.upper().str.strip()
    + '-' + hc['title'].fillna('').str.upper().str.strip()
)

nyt['book_id'] = (
    nyt['author'].fillna('').str.upper().str.strip()
    + '-' + nyt['title'].fillna('').str.upper().str.strip()
)

# Test to see if any book_ids match between the two datasets
matching_book_ids = set(hc['book_id']).intersection(set(nyt['book_id']))
print(f"Number of matching book_ids: {len(matching_book_ids)}")
```

```python
# Create new file with boolean column indicating matches

merged_df_path = '../data/hc_with_nyt_matches.csv'

# make 1 or 0 column
hc['is_in_nyt'] = hc['book_id'].apply(lambda x: 1 if x in matching_book_ids else 0)
hc.to_csv(merged_df_path, index=False)
```
